```python
Q5. What is inheritance? Give an example for each type of inheritance.

Answer-- âœ… Inheritance:

         Inheritance allows a child class to acquire properties and methods of a parent class.

     ðŸ”¹ 1. Single Inheritance

           class Parent:
                def show(self):
                    print("Parent class")

               class Child(Parent):
                    pass

              c = Child()
              c.show()

    ðŸ”¹ 2. Multiple Inheritance
             class A:
                def a(self):
                    print("Class A")

            class B:
                def b(self):
                    print("Class B")

            class C(A, B):
                pass

            obj = C()
            obj.a()
            obj.b()
     ðŸ”¹ 3. Multilevel Inheritance
           class Grandparent:
                def gp(self):
                    print("Grandparent")

            class Parent(Grandparent):
                pass

            class Child(Parent):
                pass

            c = Child()
            c.gp()
    ðŸ”¹ 4. Hierarchical Inheritance
        class Parent:
            def show(self):
                print("Parent")

        class Child1(Parent):
            pass

        class Child2(Parent):
            pass
    ðŸ”¹ 5. Hybrid Inheritance
        A combination of two or more inheritance types
        (achieved using multiple inheritance)
```
